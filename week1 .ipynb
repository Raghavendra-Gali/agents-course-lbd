```python
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role":"user", "content":"What is 2+2?"}
]
```

```python
from openai import OpenAI
gemini_chat = OpenAI(api_key=api_key,base_url=google_base_url)
# print(gemini_chat.models.list())

model_name_25_flash = "gemini-2.5-flash"

```

```python
response = gemini_chat.chat.completions.create(
    model=model_name_25_flash,
    messages=messages
)

print(response.choices[0].message.content)
```

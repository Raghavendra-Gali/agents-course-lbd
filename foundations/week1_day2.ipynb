```python
import os
from dotenv import load_dotenv
load_dotenv()
import openai
from openai import OpenAI
```

```python
gemini_api_key = os.getenv("GEMINI_API_KEY")
pptx_api_key = os.getenv("PERPLEXITY_API_KEY")
google_base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
pptx_base_url = "https://api.perplexity.ai"

gemini_client = OpenAI(api_key=gemini_api_key,base_url=google_base_url)
pptx_client = OpenAI(api_key=pptx_api_key,base_url=pptx_base_url)
```

```python
request = "Think critically and share a hard challenge for llms or nuanced question for me to ask LLMs for evaluating them.\n"
request += "Reply with question only .\n"
request += "Answer for that question is not needed"

messages = [{"role": "user", "content": request}]
print(messages)
```

```python
# print(llm_client.models.list())
```

```python
sonar_model = "sonar"
sonar_pro_model = "sonar-pro"
```

```python
sonar_question = pptx_client.chat.completions.create(
    model=sonar_pro_model,
    messages=messages
)
```

```python
from IPython.display import display, Markdown
display(Markdown(sonar_question.choices[0].message.content))
```

```python
# for model in text_models:
#     print(model)
small_models = ["gemini-2.5-flash-lite","gemini-2.5-flash"],["sonar-pro"]
```

```python
competitors = []
answers = []

for model in small_models[0]:
    try:
        resp = gemini_client.chat.completions.create(
            model=model,
            messages=[
                {"role": "user", "content": sonar_question.choices[0].message.content}
            ]
        )    
        competitors.append(model)
        answers.append(resp.choices[0].message.content)
    except Exception as e:
        print(e)
```

```python
for model in small_models[1]:
    try:
        resp = pptx_client.chat.completions.create(
            model=model,
            messages=[
                {"role": "user", "content": sonar_question.choices[0].message.content}
            ]
        )    
        competitors.append(model)
        answers.append(resp.choices[0].message.content)
    except Exception as e:
        print(e)
```

```python
print(competitors)
print(answers)
```

```python
total_responses = ""
for index,answer in enumerate(answers):
    total_responses += f"Response from LLM model {competitors[index]} is as follows : \n {answer}\n\n"
```

```python
print(total_responses)
```
